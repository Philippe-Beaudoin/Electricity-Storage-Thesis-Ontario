```python
# data preparation 
from numpy import array
import numpy as np  
import numpy.matlib
import pandas as pd
import time
import datetime
import csv
```

```python
# Weather data come from:
# https://climate.weather.gc.ca/climate_data/hourly_data_e.html?hlyRange=2002-06-04%7C2021-01-05&dlyRange=2002-06-04%7C2021-01-05&mlyRange=2003-07-01%7C2006-12-01&StationID=31688&Prov=ON&urlExtension=_e.html&searchType=stnProv&optLimit=yearRange&StartYear=2015&EndYear=2021&selRowPerPage=100&Line=255&lstProvince=ON&timeframe=1&Year=2019&Month=12&Day=1#
```

```python
file1 = 'w_2015.csv'
file2 = 'w_2016.csv'
file3 = 'w_2017.csv'
file4 = 'w_2018.csv'
file5 = 'w_2019.csv'

df1 = pd.read_csv(file1, sep=',', quoting=csv.QUOTE_NONE)
df2 = pd.read_csv(file2, sep=',', quoting=csv.QUOTE_NONE)
df3 = pd.read_csv(file3, sep=',', quoting=csv.QUOTE_NONE)
df4 = pd.read_csv(file4, sep=',', quoting=csv.QUOTE_NONE)
df5 = pd.read_csv(file5, sep=',', quoting=csv.QUOTE_NONE)

df_2015_2019 = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
df = df_2015_2019 
df
```

```python
# Keep only columns of interest
cols_of_interest = ['"""""Date/Time"""""', '"""""Temp (°C)"""""',
                    '"""""Stn Press (kPa)"""""']
df = df[cols_of_interest]

# Rename columns to get rid of quotechar
df.rename(columns = {'"""""Date/Time"""""':'Time', '"""""Temp (°C)"""""':'Temp(C)', 
                     '"""""Stn Press (kPa)"""""':'Stn_Press(kPa)'}, inplace = True)

def remove_punctuation(x):
    try:
        x = x.str.replace('"','')
    except:
        pass
    return x

df = df.apply(remove_punctuation)
```
